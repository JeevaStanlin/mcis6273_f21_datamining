In section 3.7 “Polarisation and Social Manipulabilty”, the authors suggest that “Attracting as many users as possible behind an ethical agenda seems easier if the stress is put on a catchy buzzword” . . . and “. . . we need stakeholders who truly commit to their values and promote their intrepretation and implementation”. The authors sense that, rather than interpretation of the values mentioned as a buzzword in a product, attracting more users using the buzzword for name sake is another challenge for effective Food Recommendation Systems. The emerging blockchain technology might help in maintaining the transparency of the food system.
For example, if a consumer wants to buy organic tomato ketchup, they rely on the label claim. The manufacturer of ketchup relies on the supplier of tomato puree or concentrate processor. The tomato processor relies on the claim/organic certification furnished by the farmer or farmer association. Blockchain provides a secure environment where each of the participants in the blockchain network has access to each data and these data, once entered and verified, cannot be modified. 8. List the product assessment criticisms the authors lodge against the Yuka application. Why do they feel the inclusion of “organic” as a criterion for recommendation is problematic? </br>
Yuka is a food recommender app. Yuka relies on three elements for their scoring : (i) the score a product gets from Nutriscore [12], (ii) the presence of additives in the products, and (iii) if the product is organic. Yuka automatically gives an extra 10% score to any organic product despite the fact that the positive effects of organic products on the human health are still subject to scientific debate. Yuka also make their recommendation based on a combined ratio of these factors,  60% for Nutriscore, 30% for the existence of certain additives and 10% if the product is organic. According to the authors, there are more important things to be made clear for the user, who are the natural persons behind this calculation? What makes them reliable? Are they qualified to make such an assessment?

9. Why are the crowdsourced databases that Yuka rely on a problem for the authors? </br>
Crowdsourced databases might have erroneous and outdated input. Manufacturers are allowed to change the ingredients of the products without necessarily changing the bar code. The information is not legitimate and there is no accountability and transparency in crowdsourced databases. That becomes a problem of ethical values in food recommendation.

10. Explainability and Fairness are two issues that the authors criticize Yuka heavily for. What do you think the Yuka designers could do to address the authors’ concerns? </br>
Whenever Yuka recommends an alternative based on product availability but not based on its ranking, it lacks explainability. The app could let their users know why it recommends a product ith a low score as an alternative to a healthy product.
Lack of transparency makes it impossible to be validated, there is no fairness when it comes to recommendations (as it recommends a product with score below 50 just for product availability), and it is doubtful how the product availability is carefully calculated.
All these concerns should be addressed.
